```python
import matplotlib.pyplot as plt
from shapely.geometry import Polygon

# Define the Polygons (Example Data)
IDs = [0, 1, 2, 3]
Polygon_points = {
    0: [(0, 0), (4000, 0), (4000, -4000), (0, -4000)],
    1: [(4000, 0), (8000, 0), (8000, -4000), (4000, -4000)],
    2: [(8000, 0), (12000, 0), (12000, -4000), (8000, -4000)],
    3: [(0, -4000), (4000, -4000), (4000, -8000), (0, -8000)],
}

# Plotting Polygons
for id in IDs:
    p = Polygon(Polygon_points[id])
    x, y = p.exterior.xy
    plt.plot(x, y, color='b')

# Setting plot limits
plt.xlim(0, 12000)
plt.ylim(-10000, 0)

# Define Depots
depots = {
    'Depot 1': (3669.724770642196, -3633.027522935774),
    'Depot 2': (5834.862385321091, -6788.990825688062),
    'Depot 3': (7229.357798165125, -2311.926605504583),
    'Depot 4': (9100.917431192645, -5834.862385321091),
}

# Plotting Depots
for name, (x, y) in depots.items():
    plt.scatter(x, y, s=50, c='k')
    plt.text(x, y, name, fontsize=9, ha='right')

# Define Delivery Addresses
delivery_addresses = [
    (8000, -4000),
    (9000, -5000),
    (11000, -6000)
]

# Plotting Delivery Addresses
for i, (x, y) in enumerate(delivery_addresses):
    plt.scatter(x, y, s=100, c='g', marker='x')
    plt.text(x, y, f'Delivery {i+1}', fontsize=9, ha='right')

# Function to plot a flight path
def plot_flight_path(depot_coords, delivery_coords, color='r'):
    x_values = [depot_coords[0], delivery_coords[0]]
    y_values = [depot_coords[1], delivery_coords[1]]
    plt.plot(x_values, y_values, color=color, linestyle='--')

# Plotting flight paths
plot_flight_path(depots['Depot 1'], delivery_addresses[0])
plot_flight_path(depots['Depot 2'], delivery_addresses[1])
plot_flight_path(depots['Depot 3'], delivery_addresses[2])

# Setting the aspect ratio
plt.gca().set_aspect('equal', adjustable='box')

# Display the plot
plt.show()
```
